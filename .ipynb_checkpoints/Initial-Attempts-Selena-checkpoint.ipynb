```python
#Load data
import pandas as pd
data = pd.read_csv("data.csv",header = 0)
data.head()
```

```python
#Clean up data
data = data.drop('id', 1)
data = data.drop("Unnamed: 32",1)
#In diagnosis, map M to 1 (positive) and B to 0 (negative)
data['diagnosis'] = data['diagnosis'].map({'M' : 1, 'B' : 0})
data.head()
```

```python
#Organizing data in numpy arrays
np.set_printoptions(suppress=True) #Scientific notation is annoying...
split_index = int(data.shape[0]*0.8) #Splitting test (20%) and train(80%) at this index
np_data = np.array(data, dtype=float)
#Features
x = np_data[:,1:31]
x_train = x[0:split_index,:]
x_test = x[split_index:,:]
#Target values
y = np_data[:,0]
y_train = y[:split_index]
y_test = y[split_index:]
```

```python
#Choose N

labelfont = {
        'family' : 'sans-serif',  # (cursive, fantasy, monospace, serif)
        'color'  : 'black',       # html hex or colour name
        'weight' : 'normal',      # (normal, bold, bolder, lighter)
        'size'   : 14,            # default value:12
        }

titlefont = {
        'family' : 'serif',
        'color'  : 'black',
        'weight' : 'bold',
        'size'   : 16,
        }

# how recalls fluctuate with N
recall_list=[]
for i in range(1,100):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(x_train, y_train)
    newrecall=recall_score(y_test, neigh.predict(x_test))
    recall_list.append(newrecall)
plt.plot(range(1,100),recall_list )

plt.title('Recall Function', fontdict=titlefont) 
plt.xlabel('N', fontdict=labelfont)
plt.ylabel('Recall(%)', fontdict=labelfont)

plt.show()

# Accuracy fluctuates with N
score_list=[]
for i in range(1,100):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(x_train, y_train)
    newscore=neigh.score(x_test,y_test)
    score_list.append(newscore)
plt.plot(range(1,100),score_list )

plt.title('Accuracy Function', fontdict=titlefont) 
plt.xlabel('N', fontdict=labelfont)
plt.ylabel('Accuracy(%)', fontdict=labelfont)

plt.show()

#Precision fluctuates with N 
precision_list=[]
for i in range(1,100):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(x_train, y_train)
    newprecision=precision_score(y_test, neigh.predict(x_test))
    precision_list.append(newprecision)
plt.plot(range(1,100),precision_list )

plt.title('Precision Function', fontdict=titlefont) 
plt.xlabel('N', fontdict=labelfont)
plt.ylabel('Precision(%)', fontdict=labelfont)

plt.show()

# Assessing the classifier 
neigh = KNeighborsClassifier(n_neighbors=50)
neigh.fit(x_train, y_train)
confusion_knn = confusion_matrix(neigh.predict(x_test), y_test)
print confusion_knn

   
```
